1) Display the number of attributes available in the dataset (exam_dataset.csv)?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
num_attributes = df.shape[1]
print(f'The number of attributes in the dataset is: {num_attributes}')
```

```python
num_dimensions = df.values.ndim
print(f'Number of dimension: {num_dimensions}')
```

3) Display the average of these attributes: ‘Age’, ‘Monthly Income’ and ‘Years at

```python
attributes_to_average = ['Age', 'MonthlyIncome', 'YearsAtCompany']

for attribute in attributes_to_average:
    avg = df[attribute].mean().round(2)
    print(f'Average {attribute}: {avg}')
```

4) Find the minimum and maximum ‘Monthly Income’

```python
min_income = df['MonthlyIncome'].min()
max_income = df['MonthlyIncome'].max()

print(f'Minimum Monthly Income: {min_income}')
print(f'Maximum Monthly Income: {max_income}')
```

5) Provide graphical visualization by plotting Histogram of ‘Monthly Income vs staff numbers’

```python
staff_counts = df['EmployeeCount'].value_counts()

print(staff_counts)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['MonthlyIncome'], bins=20, color='blue', edgecolor='black')
plt.xlabel('MonthlyIncome')
plt.ylabel('StaffNumbers')
plt.title('Histogram: Monthly Income vs Staff Numbers')
plt.show()
```

6) Provide graphical Visualization of the distribution between ‘Year at Company’ and ‘Monthly Income’ using the scatter plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['YearsAtCompany'], df['MonthlyIncome'], color='blue', alpha=0.7)
plt.xlabel('YearsAtCompany')
plt.ylabel('MonthlyIncome')
plt.title('Scatter Plot: Years at Company vs Monthly Income')
plt.grid(True)
plt.show()
```

7) Find the correlation between ‘Years at Company’ and ‘Monthly Income’

```python
correlation = df['YearsAtCompany'].corr(df['MonthlyIncome'])

print(f"Correlation between Years at Company and Monthly Income: {correlation}")
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='YearsAtCompany', y='MonthlyIncome', data=df)
plt.xlabel('YearsAtCompany')
plt.ylabel('MonthlyIncome')
plt.title('Scatter Plot with Trendline: Years at Company vs Monthly Income')
plt.show()
```
